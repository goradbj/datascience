#### 1. Line Plot

#### Lets consider 2 list days and temperature

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y,marker='+',color='red',markersize=16)
plt.xlabel(" Day ")
plt.ylabel("Temperature")
plt.title("Day wise temperature")
```

```python
plt.plot(x,y,marker='+',color='purple',markersize=12,linestyle="dashed",linewidth=4)
plt.xlabel(" Day ")
plt.ylabel("Temperature")
plt.title("Day wise temperature")
```

##### Format string in plot

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'+r--')
```

```python
# alpha is used for transpareny
plt.plot(x,y,marker='+',alpha=0.5)
plt.xlabel(" Day ")
plt.ylabel("Temperature")
plt.title("Day wise temperature")
```

##### Axis, grid and legend

```python
# Let us consider 4 list, days, max temp, min temp and avg temp
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.xlabel("Days")
plt.ylabel("Temp")
plt.title("Days Vs temp")
plt.legend(loc='best')
```

```python
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.xlabel("Days")
plt.ylabel("Temp")
plt.title("Days Vs temp")
plt.legend(loc='upper right')
plt.grid()
plt.savefig("image.png")
```

##### 3. Bar Chart

- bar chart is basically used for see the comparision between various parameters

```python
company=["TCS","Infy","LTI","PSL"]
revenue=[400,370,90,160]
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.bar(company,revenue,label="Revenue")
plt.xlabel("Company")
plt.ylabel("Revenue in Bill. Dollars")
plt.title("Graph for company vs Revenue")
plt.legend()
```

```python
profit=[100,150,60,90]
```

```python
plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="profit")
#plt.plot(company,revenue,label="lineshow", color="green")
#plt.scatter(company,revenue,color="black")
plt.xlabel("Company")
plt.ylabel("Revenue in Bill. Dollars")
plt.title("Graph for company vs Revenue")
plt.legend()
```

```python
plt.barh(company,revenue,label="Revenue")
plt.xlabel("Company")
plt.ylabel("Revenue in Bill. Dollars")
plt.title("Graph for company vs Revenue")
plt.legend()
```

```python
plt.barh(company,revenue,label="Revenue")
plt.barh(company,profit,label="profit")
#plt.plot(company,revenue,label="lineshow", color="green")
#plt.scatter(company,revenue,color="black")
plt.xlabel("Company")
plt.ylabel("Revenue in Bill. Dollars")
plt.title("Graph for company vs Revenue")
plt.legend()
```

- plt.bar prints vertical bar graph where as plt.barh draws horizontal graph

##### 4. Pie Charts

- it used to plot the graph with percentages for each criteria. for ex company expenses

```python
company_expenses=[100,60,40,20,80]
expense_data=["emp_sal","IF Devlopment","Food Expenses","Travelling Allow","prizes"]
```

```python
plt.pie(company_expenses,labels=expense_data,startangle=90, shadow=True,autopct='%1.2f%%',radius=1.5,explode=[0,0,0.1,0.2,0.2])
plt.savefig("C:/Users/Administrator/Desktop/TCE_expenses.png")
```

##### 5. Histogram - it is used to plot freequency

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 84, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
blood_sugar_child=[88, 93, 115, 135, 80, 84]

plt.hist([blood_sugar_men,blood_sugar_women,blood_sugar_child], 
         rwidth=1, color=['#FF6C33','blue','#0B7E0E'],label=['men','women','child'])

plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.legend()
```
